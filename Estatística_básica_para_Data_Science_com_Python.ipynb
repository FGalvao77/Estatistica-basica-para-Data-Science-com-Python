**Tendência central**

Em [estatística](https://pt.wikipedia.org/wiki/Estat%C3%ADstica), uma **tendência central** (ou, normalmente, uma **medida de tendência central**) é um valor central ou valor típico para uma [distribuição de probabilidade](https://pt.wikipedia.org/wiki/Distribui%C3%A7%C3%A3o_de_probabilidade). As medidas de tendência central mais comuns são a [média aritmética](https://pt.wikipedia.org/wiki/M%C3%A9dia_aritm%C3%A9tica), a [mediana](https://pt.wikipedia.org/wiki/Mediana_(estat%C3%ADstica)) e [moda](https://pt.wikipedia.org/wiki/Moda_(estat%C3%ADstica)). Tendências centrais podem ser calculadas tanto para um número finito de valores quanto para uma distribuição teórica, a exemplo da [distribuição normal](https://pt.wikipedia.org/wiki/Distribui%C3%A7%C3%A3o_normal). Ocasionalmente autores usam tendência central (ou **centralidade**), significando "a tendência de dados quantitativos de se agruparem ao redor de um valor central." Tal significado pode ser esperado da definição usual das palavras [tendência](https://pt.wiktionary.org/wiki/tend%C3%AAncia) e [centralidade](https://pt.wiktionary.org/wiki/centralidade) no dicionário. Autores podem julgar se dados têm tendência central forte ou fraca se baseando na [dispersão estatística](https://pt.wikipedia.org/wiki/Dispers%C3%A3o_estat%C3%ADstica), medida pelo desvio padrão ou algo similar.

```python
# importando a biblioteca para análise e manipulação de dados
import pandas as pd
```

```python
# instanciando dois dicionário com dados fictícios de 13 colaboradores de duas empresas - cargos e salários

empresa_1 = {
    'assitente_1': 1_000,
    'assistente_2': 1_000,
    'assistente_3': 1_000,
    'analista_1': 1_500,
    'analista_2': 1_520,
    'analista_3': 1_600,
    'analista_4': 1_670,
    'coordenador_1': 2_640,
    'coordenador_2': 2_730,
    'coordenador_3': 2_785,
    'gerente_1': 5_000,
    'gerente_2': 5_000,
    'ceo': 57_000
}

empresa_2 = {
    'assitente_1': 3_000,
    'assistente_2': 3_000,
    'assistente_3': 3_000,
    'analista_1': 4_200,
    'analista_2': 4_200,
    'analista_3': 4_200,
    'analista_4': 4_200,
    'coordenador_1': 6_000,
    'coordenador_2': 6_000,
    'coordenador_3': 6_000,
    'gerente_1': 8_000,
    'gerente_2': 8_000,
    'ceo': 15_000
}
```

```python
# criando dois dataframes com os dados dos dicionários acima - com colunas de cargo e salário
df_empresa_1 = pd.DataFrame(empresa_1.items(), columns=['cargo', 'salário'])
df_empresa_2 = pd.DataFrame(empresa_2.items(), columns=['cargo', 'salário'])
```

```python
# visualizando os dataframes criados
display(
    df_empresa_1,
    df_empresa_2
)
```

**Média**


Seja _n_ o número total de valores e ${x_i}$ cada valor, em que ${i=1,...,n}$. Média aritmética é a soma dos valores ${x_i}$ dividido pelo número total de valores ${n}$:

<font size=5>$\bar{x}=\frac{x_1+x_2+...+x_n}{n}=\frac{1}{n}\sum\limits^{n}_{i=1}{x_i}$</font>

```python
# visualizando a média do salario das "empresa_1" e "empresa_2"
media_empresa_1 = df_empresa_1['salário'].mean()
media_empresa_2 = df_empresa_2['salário'].mean()

print(
    '\tMédia salarial',
    f'empresa_1: {round(media_empresa_1, 2)}',
    f'empresa_2: {round(media_empresa_2, 2)}',
    sep='\n'
)
```

**Mediana**

```python
# visualizando a mediana do salario das "empresa_1" e "empresa_2"
mediana_empresa_1 = df_empresa_1['salário'].median()
mediana_empresa_2 = df_empresa_2['salário'].median()

print(
    '\tMediana salarial',
    f'empresa_1: {round(mediana_empresa_1, 2)}',
    f'empresa_2: {round(mediana_empresa_2, 2)}',
    sep='\n'
)
```

**Moda**

```python
# visualizando a moda do salario das "empresa_1" e "empresa_2"
moda_empresa_1 = df_empresa_1['salário'].mode()[0]
moda_empresa_2 = df_empresa_2['salário'].mode()[0]

print(
    '\tModa salarial',
    f'empresa_1: {round(moda_empresa_1, 2)}',
    f'empresa_2: {round(moda_empresa_2, 2)}',
    sep='\n'
)
```

**Novos dados**

```python
# instanciado um dicionário com novos dados e transformando-o em um dataframe "df"
dados = {
    'qtde': [2, 7, 9, 9, 15, 25, 53, 70, 100]
}

df = pd.DataFrame(dados)
df
```

```python
# resumo estatístico
df.describe().T
```

```python
# visualizando a média, mediana, moda e o desvio padrão do "df"
media = df['qtde'].mean()
mediana = df['qtde'].median()
moda = df['qtde'].mode()[0]
desvio_padrao = df['qtde'].std()

# print(
#     '\tResumo estatístico',
#     f'Média: {round(media, 2)}',
#     f'Mediana: {mediana}',
#     f'Moda: {moda}',
#     f'Desvio padrão: {round(desvio_padrao, 2)}',
#     sep='\n'
# )
```

```python
# plotando gráfico de histograma com a média, mediana e moda
plt.figure(figsize=(10, 5))
plt.hist(df['qtde'])
plt.axvline(x=media, color='r', linestyle='-')
plt.axvline(x=mediana, color='g', linestyle='-.')
plt.axvline(x=moda, color='black', linestyle='-')
plt.legend(['média', 'mediana', 'moda'])
plt.show()

print(
    '\n\tResumo estatístico',
    f'Média: {round(media, 2)}',
    f'Mediana: {mediana}',
    f'Moda: {moda}',
    f'Desvio padrão: {round(desvio_padrao, 2)}',
    sep='\n'
)
```

```python
# plotando gráfico de pontos com a linha da média dos dados
x = range(0, len(dados['qtde']))

plt.figure(figsize=(10, 5))
plt.scatter(x, dados['qtde'])
plt.axhline(y=media, color='r', linestyle='-')
plt.legend(['média', 'dados'])
plt.show()
```
